```python
def calcular_distancia_euclidiana(filme1, filme2, matriz_usuario_filme):
    # Obter as avaliaÃ§Ãµes do filme1
    avaliacoes_filme1 = matriz_usuario_filme.loc[filme1]

    # Obter as avaliaÃ§Ãµes do filme2
    avaliacoes_filme2 = matriz_usuario_filme.loc[filme2]

    # Calcular a distÃ¢ncia euclidiana
    diferenca_quadratica = np.sum((avaliacoes_filme1 - avaliacoes_filme2)**2)
    return np.sqrt(diferenca_quadratica)

def recomendar_filmes_por_titulo(titulo_filme, matriz_usuario_filme, dados, num_recomendacoes=3):
    # Calcular a distÃ¢ncia euclidiana entre o tÃ­tulo de filme e todos os outros filmes
    distancias = {}
    for filme in matriz_usuario_filme.index:
        if filme != titulo_filme:
            distancias[filme] = calcular_distancia_euclidiana(titulo_filme, filme, matriz_usuario_filme)
    
    # Ordenar os filmes por distÃ¢ncia (quanto menor a distÃ¢ncia, mais similar)
    filmes_similares = sorted(distancias.items(), key=lambda x: x[1])
    
    # Remover os filmes repetidos da lista de recomendaÃ§Ãµes
    recomendacoes = set(filmes_similares)
    
    # Retornar os 3 filmes mais bem avaliados
    return list(recomendacoes)[:num_recomendacoes]
```

```python
# Criar uma matriz de usuÃ¡rios x filmes
matriz_usuario_filme = avaliacoes_e_filmes.pivot_table(columns = 'ID_USUARIO', index = 'TITULO', values = 'AVALIACAO')
```

```python
# Filtrar os tÃ­tulos Ãºnicos
titulos_unicos = avaliacoes_e_filmes['TITULO'].unique()

# Criar um DataFrame com os tÃ­tulos Ãºnicos
df_titulos_unicos = pd.DataFrame(titulos_unicos)

# Exportar o DataFrame para um arquivo Excel
df_titulos_unicos.to_excel('titulos_unicos.xlsx', index=False)

# Chamar a funÃ§Ã£o `recomendar_filmes_por_titulo()`
recomendacoes = recomendar_filmes_por_titulo('10,000 BC', matriz_usuario_filme, avaliacoes_e_filmes, num_recomendacoes=3)

# Imprimir as recomendaÃ§Ãµes
print(recomendacoes)
```
