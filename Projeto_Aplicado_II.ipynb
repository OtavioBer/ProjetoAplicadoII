##Analise dos dados de classificaes dos filmes - DF2

```python
notas = pd.read_csv("ratings_small.csv")
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId").mean()["rating"]
nota_media_por_filme.head()
```

```python
ax = sns.distplot(nota_media_por_filme.values)
ax.set(xlabel="Nota mdia", ylabel="Densidade")
ax.set_title("Mdia de votos em filmes")
```

```python
quantidade_de_votos_por_filme = notas.groupby("movieId").count()
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query("rating >= 10").index
filmes_com_pelo_menos_10_votos.values
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel="Nota mdia", ylabel="Densidade")
ax.set_title("Mdia de votos em filmes com 10 ou mais votos")
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos.values,
                          hist_kws = {'cumulative':True},
                          kde_kws = {'cumulative':True})
ax.set(xlabel="Nota mdia", ylabel="% acumulada de filmes")
ax.set_title("Mdia de votos em filmes com 10 ou mais votos")
```

```python
ax = sns.boxplot(x=nota_media_dos_filmes_com_pelo_menos_10_votos.values)
ax.set(xlabel="Nota mdia do filme")
ax.set_title("Distribuio de nota mdia dos filmes, dentre os filmes com 10 ou mais votos")
```

Mdia dos filmes com pelo menos 10 votos - Classificao

```python
print("Mdia dos filmes com pelo menos 10 votos", nota_media_dos_filmes_com_pelo_menos_10_votos.mean())
```

```python
#selecionar amostras aleatrias
np.random.seed(2642)
temp = nota_media_dos_filmes_com_pelo_menos_10_votos.sample(frac=1)

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```

##Teste da amostra - Intervalo de confiana da mdia com t e ztest

Analisaremos qual o intervalo de confiana que temos para as mdias dos nossos filmes. Isso no quer dizer que todos os filmes do mundo tero uma mdia 3.47, que  a mdia dos filmes da base de dados analisada que apresentam pelo menos 10 votos, mas queremos criar uma generalizao, a partir dessa amostra, sobre os filmes que existem no mundo.

https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.zconfint.html

```python
from statsmodels.stats.weightstats import zconfint

zconfint(nota_media_dos_filmes_com_pelo_menos_10_votos)
```
